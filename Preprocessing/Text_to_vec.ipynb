```python
glove_vectors = gensim.downloader.load('glove-twitter-50')
```

```python
def glove_transformer(X):
    X = X.copy()
    X = X.join(X.apply(
        lambda x: 
            pd.Series(sum([
                glove_vectors[word] 
                for word in x["question1"].split() 
                if word in glove_vectors.vocab
            ]), index=["glv_q1_"+str(i) for i in range(50)]), axis=1, result_type="expand"))
    X = X.join(X.apply(
        lambda x: 
            pd.Series(sum([
                glove_vectors[word] 
                for word in x["question2"].split() 
                if word in glove_vectors.vocab
            ]), index=["glv_q2_"+str(i) for i in range(50)]), axis=1, result_type="expand"))
    return X
```

```python
df_train_lemma.question1=df_train_lemma.question1.astype('U')
df_test_lemma.question1=df_test_lemma.question1.astype('U')
df_train_lemma.question2=df_train_lemma.question2.astype('U')
df_test_lemma.question2=df_test_lemma.question2.astype('U')
```
